Projeto de ETL para o bootcamp Santander

## 1. Extract — Leitura da Base de Dados Bruta

```python
import pandas as pd

# Lendo dados do arquivo CSV
df_raw = pd.read_csv("filmes.csv")

df_raw
```

## 2. Transform — Padronização, Enriquecimento e Regras de Negócio

```python
df_transformed = df_raw.copy()

df_transformed["genre"] = df_transformed["genre"].fillna("Not Informed")

def classify_movie(r):
    if r is None or pd.isna(r):
        return "No Rating"
    elif r >= 8.5:
        return "Excellent"
    elif r >= 7:
        return "Good"
    else:
        return "Average"

df_transformed["category"] = df_transformed["rating"].apply(classify_movie)

df_transformed["slug"] = df_transformed["title"].str.lower().str.replace(" ", "-")

df_transformed
```

## 3. Load — Salvando o Resultado em CSV

```python
df_transformed.to_csv("movies_processed.csv", index=False)
df_transformed
```
